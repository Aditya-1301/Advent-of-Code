```python
import re
```

```python
f = open("input3.txt")
pattern = r'mul\((-?\d+),\s*(-?\d+)\)'
result = 0
for i in f:
    muls = re.findall(pattern, i)
    result += sum([int(i1) * int(i2) for (i1, i2) in muls])
result
```

### NOTE:
It's necessary to include do() at the beginning of the input for the next part to work

```python
f = open("input3.txt")
input = []
for i in f:
    input.extend(i.split("do"))
print(input)
i1 = []
for i in input:
    if i.startswith("n't()"):
        continue
    else:
        i1.extend(i)
input= "".join(i1)
input
```

```python
pattern = r'mul\((-?\d+),\s*(-?\d+)\)'
result = 0
muls = re.findall(pattern, input)
print(muls)
result += sum([int(i1) * int(i2) for (i1, i2) in muls])
result
```
